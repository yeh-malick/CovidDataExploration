```python
# lIBRARIES set up

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams ['figure.figsize'] = (12,8)


df = pd.read_csv(r'C:\Users\USER\Desktop\PortfolioProjetMovies\movies.csv')
```

```python
df.head()
```

```python
#Checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
df.dtypes
```

```python
 #data type change
df['budget'] =df ['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Fixing year column

df['correctyear'] = df['released'].astype(str).str[:4]
df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Getting unique list of companies

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Deleting duplicates

df.drop_duplicates()
```

```python
# Hypotheses: Budget, Genre, score, director influence gross revenue


#Scatter plot hypothesis 1 - Budget x Gross revenue

plt.scatter(x=df['budget'],y=df['gross'])
plt.show()
```

```python
# Plot budget x gross with linear regression

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Correlation matrix

df.corr() 
#df.corr(method='pearson/kendall/spearman')
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix')
plt.xlabel('Movie details')
plt.ylabel('Movie details')
plt.show()
```

```python
# Plot score x gross with linear regression

sns.regplot(x='score', y='gross', data=df, scatter_kws={"color":"green"}, line_kws={"color":"blue"})
```

```python
# Plot votes x gross with linear regression --> popularity

sns.regplot(x='votes', y='gross', data=df, scatter_kws={"color":"yellow"}, line_kws={"color":"blue"})
```

```python
# Plot runtime x gross with linear regression

sns.regplot(x='runtime', y='gross', data=df, scatter_kws={"color":"orange"}, line_kws={"color":"blue"})
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'): 
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name]= df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix')
plt.xlabel('Movie details')
plt.ylabel('Movie details')
plt.show()
```
